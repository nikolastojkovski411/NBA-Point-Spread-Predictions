```python
# Convert "Last, First" -> "First Last"
def format_name(name):
    if ',' in name:
        last, first = name.split(',', 1)
        return first.strip() + ' ' + last.strip()
    return name.strip()

missed_df['PLAYER'] = missed_df['PLAYER'].apply(format_name)
```

```python
import pandas as pd
from collections import defaultdict
import numpy as np

# Ensure datetime
missed_df['DATE'] = pd.to_datetime(missed_df['DATE'])
boxscores_df['GAME_DATE'] = pd.to_datetime(boxscores_df['GAME_DATE'])

# Sort and prepare dictionary
boxscores_df = boxscores_df.sort_values(by=['PLAYER_NAME', 'GAME_DATE'])
player_games = defaultdict(pd.DataFrame)
for player_id in boxscores_df['PLAYER_NAME'].unique():
    player_games[player_id] = boxscores_df[boxscores_df['PLAYER_NAME'] == player_id]

```

```python
def get_minute_weighted_bpm(player_id, missed_date):
    if player_id not in player_games:
        return None
    games = player_games[player_id]
    prior_games = games[games['GAME_DATE'] < missed_date]
    last_n = prior_games.tail(10)
    if last_n.empty:
        return None
    
    plus_minus = last_n['PLUS_MINUS'].values
    minutes = last_n['MIN'].astype(float).values  # make sure it's float for weighting
    
    if minutes.sum() == 0:
        return 0  # Avoid division by zero
    
    weighted_avg = np.average(plus_minus, weights=minutes)
    return round(weighted_avg, 2)

# Apply to each row
missed_df['rolling_bpm_5'] = missed_df.apply(
    lambda row: get_minute_weighted_bpm(row['PLAYER'], row['DATE']),
    axis=1
)
```

```python
# 1 = definitely out, 0 = definitely playing
status_weight = {
    'Out': 1.0,
    'Doubtful': 0.75,
    'Questionable': 0.5,
    'Probable': 0.25,
    'Available': 0.0
}

missed_df['injury_weight'] = missed_df['STATUS'].map(status_weight)
```

```python
# Weighted minutes-adjusted box +/- by injury likelihood
missed_df['injury_bpm_impact'] = missed_df['rolling_bpm_5'] * missed_df['injury_weight']
```

```python
missed_df
```

```python
team_injury_impact = (
    missed_df.groupby(['TEAM', 'DATE', 'GAME'])['injury_weight']
    .sum()
    .reset_index()
    .rename(columns={'injury_weight': 'team_missing_bpm'})
)
```

```python
#team_injury_impact['GAME_formatted'] = 
team_injury_impact['GAME'] = team_injury_impact['GAME'].apply(
    lambda x: f"{x.split('@')[1].strip()} vs. {x.split('@')[0].strip()}"
)
```
