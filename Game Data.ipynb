```python
from nba_api.stats.static import teams
from nba_api.stats.endpoints import teamgamelog
from nba_api.stats.library.parameters import SeasonAll
import pandas as pd
import time
```

```python
team_dict = teams.get_teams()
all_teams_df = pd.DataFrame()
for team in team_dict:
    team_id = team['id']
    gamelog = teamgamelog.TeamGameLog(team_id=team_id, season = SeasonAll.all)
    df = gamelog.get_data_frames()[0]
    print(df)

    df['TEAM_NAME'] = team['full_name']
    df = df.reindex(columns=['SEASON_ID','Game_ID', 'TEAM_NAME'] + [col for col in df.columns if col != 'SEASON_ID' and col != 'Game_ID' and col != 'TEAM_NAME'])

    df = df.drop(['Team_ID'], axis=1)
    
    df['Season'] = df['GAME_DATE'].str.rsplit(',').str[-1] 
    df = df.reindex(columns=['Season',] + [col for col in df.columns if col != 'Season'])
    df = df.drop(['SEASON_ID'], axis=1)
    all_teams_df = pd.concat([all_teams_df, df], ignore_index=True)
    time.sleep(3)  ##rate limit to 20 requests every minute
    print("Finished with:"+ team['full_name'])
```

```python
##after the df is filled I change the Season column from type object to type int64
all_teams_df[["Season"]] = all_teams_df[["Season"]].apply(pd.to_numeric)
##after which I sort the dataframe by season and then by player names in alphabetical order
all_teams_df = all_teams_df.sort_values(by=['Season', 'TEAM_NAME'], ascending=[True, True])
```

```python
# Read the data and set proper index if there's an unnamed column
all_teams_df = pd.read_csv(r'C:\Users\ninuy\DSC 672\data\NBA Team Box Score Stats.csv', index_col=0)

# Columns that should not be renamed (i.e., identifiers)
id_cols = ['Season', 'Game_ID', 'TEAM_NAME', 'GAME_DATE', 'MATCHUP', 'WL']

cols_to_drop = ['W', 'L', 'W_PCT']
all_teams_df.drop(columns=cols_to_drop, inplace=True)


# Filter for home and away games using the "@" in the MATCHUP column
home_teams_df = all_teams_df[~all_teams_df['MATCHUP'].str.contains("@")].copy()
away_teams_df = all_teams_df[all_teams_df['MATCHUP'].str.contains("@")].copy()

# Rename columns in each DataFrame
home_teams_df.rename(
    columns={col: f'HOME_{col}' for col in home_teams_df.columns if col not in id_cols},
    inplace=True
)
away_teams_df.rename(
    columns={col: f'AWAY_{col}' for col in away_teams_df.columns if col not in id_cols},
    inplace=True
)

# Drop duplicated ID columns from away_teams_df before merge
cols_to_drop = ['Season', 'GAME_DATE', 'MATCHUP', 'WL', 'TEAM_NAME']  # Keep 'Game_ID' for joining
away_teams_df.drop(columns=cols_to_drop, inplace=True)

# Merge on Game_ID with no suffixes
combined_df = pd.merge(home_teams_df, away_teams_df, on='Game_ID', how='inner')

# Calculate score margin (home - away)
combined_df['POINT_MARGIN'] = combined_df['HOME_PTS'] - combined_df['AWAY_PTS']

# Count rows with any missing values, grouped by season
missing_per_season = combined_df[combined_df.isna().any(axis=1)].groupby('Season').size()

print(missing_per_season)

display(combined_df)

combined_df.to_csv(r'C:\Users\ninuy\DSC 672\data\NBA Team Box Score Stats Procesed.csv', index = True)
```
