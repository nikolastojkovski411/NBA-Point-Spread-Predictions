```python
import pandas as pd
from tqdm import tqdm
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk

# Download VADER lexicon (only needed once)
nltk.download("vader_lexicon")

# Initialize VADER sentiment analyzer
vader = SentimentIntensityAnalyzer()

# Function to get VADER sentiment
def get_vader_sentiment(text):
    scores = vader.polarity_scores(text)
    return {
        "negative": scores["neg"],
        "neutral": scores["neu"],
        "positive": scores["pos"],
        "compound": scores["compound"]
    }

# Analyze a list of texts with progress bar
def analyze_texts(text_list):
    results = []
    for text in tqdm(text_list, desc="Analyzing with VADER"):
        result = get_vader_sentiment(text)
        result["text"] = text
        results.append(result)
    return pd.DataFrame(results)

# Example usage
texts = ["Amazing experience!", "It was okay.", "Worst thing ever."]
df = analyze_texts(texts)
print(df)
```
